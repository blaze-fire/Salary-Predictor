```python
import pandas as pd
import re
```

```python
df = pd.read_csv('./data/temp.csv', names=['Job_position', 'Company', 'Salary', 'requirements', 'rating'])
```

```python
df.info()
```

```python
df.drop_duplicates(ignore_index=True, inplace=True)
```

```python
df.info()
```

```python
def get_states(df):

    with open('utils/states.txt', 'r') as f:
        states = f.read()
        states_list = states.split(',')
    f.close()

    job_states = []

    for j in range(len(df)):
        counter = 0

        for i in states_list:
            if re.findall(i, df['Company'][j].lower()):
                job_states.append(i)
                if i != 'delhi':
                    df.loc[j, 'Company'] = re.sub(i, '',  df['Company'][j].lower())
                counter = 1
                break

        if counter == 0:
            job_states.append('State_missing')

    df['State'] = job_states

    return df

def replace_comp(df):

    for j in range(len(df)):

        if re.findall(r'(ltd|review|reviews|remote|temporarily|\d.?)', df['Company'][j].lower()):
            df.loc[j, 'Company'] = re.sub(r'(ltd|review|reviews|remote|temporarily|\d.?)', '', df['Company'][j].lower())

    return df

def city(df):

    with open('utils/cities.txt', 'r') as f:
        cities = f.read()
        cities_list = cities.split(',')
    f.close()

    job_cities = []

    for j in range(len(df)):
        counter = 0

        for i in cities_list:
            if re.findall(i, df['Company'][j].lower()):
                job_cities.append(i)
                df.loc[j, 'Company'] = re.sub(i, ' ', df['Company'][j].lower())
                counter = 1
                break

        if counter == 0:
            job_cities.append('city_missing')

    df['Cities'] = job_cities

    return df
```

```python
temp = replace_comp(df)
temp = city(temp)
```

```python
temp = get_states(temp)
```

```python
temp
```

```python
for i in temp.index.values:
    if temp.loc[i, 'Cities'] == 'delhi':
        temp.loc[i, 'State'] = 'delhi'
```

```python
temp
```

```python
import numpy as np
l=[]
for i in range(0, len(temp)):
    l.append(np.nan)

temp['Location'] = l
```

```python
for i in temp.index.values:
    location = ''
    
    if temp.loc[i, 'Cities'] != 'city_missing':
        location += temp.loc[i, 'Cities'] + ', '
        
    if temp.loc[i, 'State'] != 'State_missing':
        location += temp.loc[i, 'State']
    temp.loc[i, 'Location'] = location

temp.drop(['State', 'Cities'], axis=1, inplace=True)
```
